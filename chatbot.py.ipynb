```python
# 설치 필요 라이브러리
# !pip install sentence-transformers transformers faiss-cpu langchain

import os
import faiss
from tqdm import tqdm
from sentence_transformers import SentenceTransformer
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
from langchain.schema import Document

# ======================================
# 1. 청크 로드
# ======================================
chunk_dir = "chunks"
chunk_files = sorted([f for f in os.listdir(chunk_dir) if f.endswith(".txt")])
chunks = []

for fname in chunk_files:
    with open(os.path.join(chunk_dir, fname), "r", encoding="utf-8") as f:
        chunks.append(f.read())

print(f"✅ 총 {len(chunks)}개의 청크를 불러왔습니다.")

documents = [Document(page_content=chunk) for chunk in chunks]

# ======================================
# 2. 벡터 임베딩 (SentenceTransformer 사용)
# ======================================
embedding_model = SentenceTransformer("BAAI/bge-small-en")
doc_texts = [doc.page_content for doc in documents]
embeddings = embedding_model.encode(doc_texts, show_progress_bar=True)

# ======================================
# 3. FAISS 벡터 인덱스 생성
# ======================================
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(embeddings)

# ======================================
# 4. Flan-T5 모델 및 토크나이저 로딩
# ======================================
model_name = "google/flan-t5-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
llm_model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

pipe = pipeline("text2text-generation", model=llm_model, tokenizer=tokenizer, max_new_tokens=512)

# ======================================
# 5. 질문 답변 함수 정의 (RAG)
# ======================================
def ask_question(query, top_k=3):
    # query 임베딩
    query_vec = embedding_model.encode([query])
    D, I = index.search(query_vec, top_k)
    relevant_chunks = [doc_texts[i] for i in I[0]]

    # 참고 문서 내용 합치기
    context = "\n---\n".join(relevant_chunks)

    # 프롬프트 생성
    prompt = f"다음 내용을 참고하여 질문에 답변해 주세요.\n\n{context}\n\n질문: {query}\n답변:"
    
    response = pipe(prompt)[0]['generated_text']
    return response.strip()

```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, TrainingArguments, Trainer, DataCollatorForLanguageModeling, BitsAndBytesConfig
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training, TaskType
import torch
from datasets import Dataset
import os

def load_chunks(folder):
    data = []
    for fname in os.listdir(folder):
        path = os.path.join(folder, fname)
        if os.path.isdir(path):
            continue  # 폴더 무시
        if not fname.endswith(".txt"):
            continue  # .txt 파일만 대상
        with open(path, "r", encoding="utf-8") as f:
            data.append({"text": f.read()})
    return Dataset.from_list(data)


dataset = load_chunks("chunks")

# 2. 모델 및 토크나이저 로드
model_id = "EleutherAI/pythia-70m"  # 소형 모델 예시

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

model = AutoModelForCausalLM.from_pretrained(model_id, quantization_config=bnb_config, device_map="auto")
tokenizer = AutoTokenizer.from_pretrained(model_id)
tokenizer.pad_token = tokenizer.eos_token

# 3. LoRA 설정 및 모델 준비
model = prepare_model_for_kbit_training(model)

lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    lora_dropout=0.1,
    bias="none",
    task_type=TaskType.CAUSAL_LM  # Pythia는 CausalLM입니다.
    # target_modules 생략 => 자동 매칭
)
model = get_peft_model(model, lora_config)

# 4. 데이터 전처리
def tokenize(sample):
    return tokenizer(sample["text"], truncation=True, padding="max_length", max_length=512)

tokenized_dataset = dataset.map(tokenize, batched=True)
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)

# 5. 학습 설정
training_args = TrainingArguments(
    output_dir="./slm_lora",
    per_device_train_batch_size=4,
    num_train_epochs=3,
    logging_steps=10,
    save_total_limit=1,
    save_steps=100,
    gradient_checkpointing=True,
    fp16=True,
    report_to="none"
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    tokenizer=tokenizer,
    data_collator=data_collator
)

trainer.train()

model.save_pretrained("./slm_lora")
tokenizer.save_pretrained("./slm_lora")
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
import re

class SimpleRAGChatbot:
    def __init__(self, faiss_path="embed_faiss"):
        """
        간단한 RAG 챗봇 초기화 
        
        Args:
            faiss_path: FAISS 벡터 데이터베이스 경로
        """
        print("임베딩 모델 로딩 중...")
        self.embedding_model = HuggingFaceEmbeddings(
            model_name="sentence-transformers/all-MiniLM-L6-v2"
        )
        
        print("FAISS 벡터 데이터베이스 로딩 중...")
        self.db = FAISS.load_local(faiss_path, self.embedding_model, allow_dangerous_deserialization=True)
        
        print("챗봇 준비 완료!")
    
    def search_documents(self, query, k=3):
        """문서 검색"""
        docs = self.db.similarity_search(query, k=k)
        return docs
    
    def get_relevant_content(self, question, k=3):
        """질문과 관련된 내용 검색 및 정리"""
        docs = self.search_documents(question, k)
        
        if not docs:
            return "관련된 문서를 찾을 수 없습니다."
        
        # 검색된 문서들의 내용 결합
        content_parts = []
        sources = []
        
        for i, doc in enumerate(docs, 1):
            content_parts.append(f"[참고문서 {i}]\n{doc.page_content.strip()}")
            sources.append({
                'source': doc.metadata.get('source', 'Unknown'),
                'page': doc.metadata.get('page', 'Unknown'),
                'paragraph': doc.metadata.get('paragraph', 'Unknown')
            })
        
        combined_content = "\n\n".join(content_parts)
        
        return {
            'content': combined_content,
            'sources': sources,
            'doc_count': len(docs)
        }
    
    def extract_keywords(self, text):
        """텍스트에서 키워드 추출"""
        # 간단한 키워드 추출 (한글, 영문, 숫자)
        keywords = re.findall(r'[가-힣a-zA-Z0-9]+', text)
        # 길이가 2 이상인 키워드만 선택
        keywords = [k for k in keywords if len(k) >= 2]
        return list(set(keywords))  # 중복 제거
    
    def answer_question(self, question):
        """질문에 대한 답변 생성"""
        result = self.get_relevant_content(question)
        
        if isinstance(result, str):  # 문서를 찾을 수 없는 경우
            return result
        
        # 간단한 답변 형식 생성
        answer = f"""
질문: {question}

검색된 관련 내용 ({result['doc_count']}개 문서):

{result['content']}

---
참고 문서 정보:
"""
        
        for i, source in enumerate(result['sources'], 1):
            answer += f"{i}. {source['source']} (페이지: {source['page']}, 단락: {source['paragraph']})\n"
        
        return answer
    
    def chat(self):
        """대화형 챗봇 실행"""
        print("\n=== 간단한 RAG 챗봇 시작 ===")
        print("질문을 입력하세요. 종료하려면 'quit' 또는 'exit'를 입력하세요.")
        print("키워드 검색하려면 'keyword: [키워드]' 형식으로 입력하세요.")
        print("-" * 50)
        
        while True:
            question = input("\n질문: ").strip()
            
            if question.lower() in ['quit', 'exit', '종료']:
                print("챗봇을 종료합니다.")
                break
            
            if question.startswith('keyword:'):
                # 키워드 검색
                keyword = question[8:].strip()
                docs = self.search_documents(keyword, k=5)
                
                print(f"\n'{keyword}' 관련 검색 결과 ({len(docs)}개):")
                for i, doc in enumerate(docs, 1):
                    print(f"\n[문서 {i}]")
                    print(f"출처: {doc.metadata.get('source', 'Unknown')}")
                    print(f"내용 미리보기: {doc.page_content[:150]}...")
                    
                    # 키워드가 포함된 부분 강조 표시
                    if keyword.lower() in doc.page_content.lower():
                        print("✓ 키워드 포함됨")
            
            elif question:
                print("\n검색 중...")
                answer = self.answer_question(question)
                print(answer)
    
    def get_document_stats(self):
        """데이터베이스 통계 정보"""
        # 전체 문서 수 확인 (근사치)
        sample_docs = self.db.similarity_search("", k=1000)  # 최대 1000개까지 검색
        
        print(f"데이터베이스 정보:")
        print(f"- 검색 가능한 문서 수: {len(sample_docs)}개 (근사치)")
        
        if sample_docs:
            # 샘플 메타데이터 분석
            sources = set()
            pages = set()
            
            for doc in sample_docs[:100]:  # 처음 100개만 분석
                sources.add(doc.metadata.get('source', 'Unknown'))
                pages.add(doc.metadata.get('page', 'Unknown'))
            
            print(f"- 포함된 소스 파일 수: {len(sources)}개")
            print(f"- 포함된 페이지 수: {len(pages)}개")

# 사용 예시
if __name__ == "__main__":
    try:
        chatbot = SimpleRAGChatbot()
        
        # 데이터베이스 정보 출력
        chatbot.get_document_stats()
        
        # 챗봇 시작
        chatbot.chat()
        
    except KeyboardInterrupt:
        print("\n챗봇이 중단되었습니다.")
    except Exception as e:
        print(f"오류 발생: {e}")
        print("FAISS 데이터베이스가 올바르게 생성되었는지 확인해주세요.")
```

```python
import os
import re
import torch
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from peft import PeftModel, PeftConfig


class InteractiveRAGChatbot:
    def __init__(self, faiss_path="embed_faiss"):
        print("임베딩 모델 로딩 중...")
        self.embedding_model = HuggingFaceEmbeddings(
            model_name="sentence-transformers/all-MiniLM-L6-v2"
        )

        print("FAISS 벡터 데이터베이스 로딩 중...")
        self.db = FAISS.load_local(faiss_path, self.embedding_model, allow_dangerous_deserialization=True)

        print("Flan-T5 모델 + LoRA 설정 중...")
        base_model = "google/flan-t5-base"
        peft_model_path = "slm_lora"  # 학습된 LoRA 디렉토리

        self.tokenizer = AutoTokenizer.from_pretrained(base_model)
        model = AutoModelForSeq2SeqLM.from_pretrained(base_model, device_map="auto", torch_dtype=torch.float16)
        self.llm = PeftModel.from_pretrained(model, peft_model_path)
        self.llm.eval()
        print("모델 로딩 완료!")

    def extract_keywords(self, text):
        keywords = re.findall(r'[가-힣a-zA-Z0-9]+', text)
        return list(set(k for k in keywords if len(k) >= 2))

    def filter_docs_by_keywords(self, keywords):
        all_docs = self.db.similarity_search("", k=1000)
        matched = [doc for doc in all_docs if any(k in doc.page_content for k in keywords)]
        return matched

    def generate_answer(self, context, question):
        # context를 512 토큰 이하로 자름
        context = context[:1000]

        prompt = f"""
        다음 문서를 참고하여 질문에 자연스럽고 정확한 답변을 생성하세요. 3줄 이내로 간결하게 요약해주세요.

        문서 내용:
        {context}

        질문: {question}
        """

        inputs = self.tokenizer(prompt, return_tensors="pt", truncation=True, max_length=512).to(self.llm.device)
        try:
            with torch.no_grad():
                output_ids = self.llm.generate(
                    **inputs,
                    max_new_tokens=150,
                    num_beams=5,
                    do_sample=False
                )
            output = self.tokenizer.decode(output_ids[0], skip_special_tokens=True)
            return output.strip()
        except Exception as e:
            return f"⚠️ SLM 답변 생성 실패: {e}"

    def chat(self):
        print("\n=== SLM & RAG 기반 대화형 챗봇 시작 ===")
        print("문서 내용 기반 검색 및 자연어 응답을 지원합니다. '종료' 입력 시 종료됩니다.")
        print("----------------------------------------------------")

        while True:
            user_input = input("\n어떤 주제에 대해 궁금하신가요?: ").strip()

            if user_input.lower() in ['종료', 'quit', 'exit']:
                print("챗봇을 종료합니다.")
                break

            keywords = self.extract_keywords(user_input)
            matched_docs = self.filter_docs_by_keywords(keywords)

            if not matched_docs:
                print("해당 키워드와 관련된 문서를 찾을 수 없습니다. 다른 키워드를 입력해 주세요.")
                continue

            print(f"\n[{', '.join(keywords)}] 키워드가 포함된 문서 일부를 검색했습니다.")
            print("관련 내용 중에서 질문을 해보세요.")

            while True:
                followup = input("궁금한 내용을 입력해주세요: ").strip()
                if followup.lower() in ['종료', 'quit', 'exit']:
                    print("챗봇을 종료합니다.")
                    return

                print("\n문서 검색 중...")
                results = self.db.similarity_search(followup, k=3)

                if not results:
                    print("관련 내용을 찾을 수 없습니다. 다시 입력해주세요.")
                    continue

                context = "\n\n".join(doc.page_content.strip() for doc in results)
                answer = self.generate_answer(context, followup)

                print("\n--- 답변 ---")
                print(answer)
                print("\n출처:")
                for i, doc in enumerate(results, 1):
                    metadata = doc.metadata
                    print(f"- 문서{i}: {metadata.get('source', '알 수 없음')} (페이지: {metadata.get('page', '?')})")
                print("\n다른 질문이 있으시면 계속 입력해주세요. 종료하려면 '종료'를 입력하세요.")
                break


if __name__ == "__main__":
    try:
        chatbot = InteractiveRAGChatbot()
        chatbot.chat()
    except KeyboardInterrupt:
        print("\n챗봇이 중단되었습니다.")
    except Exception as e:
        print(f"오류 발생: {e}")
        print("FAISS 데이터베이스 또는 로컬 LLM 모델 구성을 확인해주세요.")
```

```python
import os
import re
import torch
import gc
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from peft import PeftModel, PeftConfig


class InteractiveRAGChatbot:
    def __init__(self, faiss_path="embed_faiss"):
        # CUDA 캐시 정리
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
            torch.cuda.synchronize()
        
        print("임베딩 모델 로딩 중...")
        self.embedding_model = HuggingFaceEmbeddings(
            model_name="sentence-transformers/all-MiniLM-L6-v2",
            model_kwargs={'device': 'cpu'}  # 임베딩은 CPU에서 실행
        )

        print("FAISS 벡터 데이터베이스 로딩 중...")
        self.db = FAISS.load_local(faiss_path, self.embedding_model, allow_dangerous_deserialization=True)

        print("Flan-T5 모델 + LoRA 설정 중...")
        base_model = "google/flan-t5-base"
        peft_model_path = "slm_lora"

        # 디바이스 설정
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        print(f"사용 디바이스: {self.device}")
        
        # 토크나이저 로딩
        self.tokenizer = AutoTokenizer.from_pretrained(base_model)
        
        # 모델 로딩 시 더 안전한 설정
        try:
            if torch.cuda.is_available():
                # GPU 메모리 상태 확인
                torch.cuda.empty_cache()
                
                # 모델 로딩 (더 보수적인 설정)
                model = AutoModelForSeq2SeqLM.from_pretrained(
                    base_model, 
                    torch_dtype=torch.float16,
                    low_cpu_mem_usage=True,
                    device_map="auto"
                )
                
                # LoRA 모델 로딩
                self.llm = PeftModel.from_pretrained(model, peft_model_path)
                
                # 명시적으로 디바이스 이동
                if not hasattr(self.llm, 'device') or self.llm.device != self.device:
                    self.llm = self.llm.to(self.device)
            else:
                # CPU 사용 시
                model = AutoModelForSeq2SeqLM.from_pretrained(
                    base_model,
                    torch_dtype=torch.float32
                )
                self.llm = PeftModel.from_pretrained(model, peft_model_path)
                self.llm = self.llm.to(self.device)
            
            self.llm.eval()
            
            # 메모리 정리
            gc.collect()
            if torch.cuda.is_available():
                torch.cuda.empty_cache()
                
        except Exception as e:
            print(f"모델 로딩 중 오류 발생: {e}")
            print("CPU 모드로 전환합니다...")
            self.device = torch.device("cpu")
            model = AutoModelForSeq2SeqLM.from_pretrained(base_model, torch_dtype=torch.float32)
            self.llm = PeftModel.from_pretrained(model, peft_model_path)
            self.llm = self.llm.to(self.device)
            self.llm.eval()
        
        print("모델 로딩 완료!")

    def extract_keywords(self, text):
        keywords = re.findall(r'[가-힣a-zA-Z0-9]+', text)
        return list(set(k for k in keywords if len(k) >= 2))

    def filter_docs_by_keywords(self, keywords):
        all_docs = self.db.similarity_search("", k=1000)
        matched = [doc for doc in all_docs if any(k in doc.page_content for k in keywords)]
        return matched

    def generate_answer(self, context, question):
        # context를 더 짧게 자름 (메모리 절약)
        context = context[:800]

        prompt = f"""
        다음 문서를 참고하여 질문에 자연스럽고 정확한 답변을 생성하세요. 3줄 이내로 간결하게 요약해주세요.

        문서 내용:
        {context}

        질문: {question}
        """

        try:
            # 토큰화 시 더 보수적인 길이 설정
            inputs = self.tokenizer(
                prompt, 
                return_tensors="pt", 
                truncation=True, 
                max_length=400,  # 더 짧게 설정
                padding=True
            )
            
            # 입력을 올바른 디바이스로 이동
            inputs = {k: v.to(self.device) for k, v in inputs.items()}
            
            # CUDA 동기화
            if torch.cuda.is_available():
                torch.cuda.synchronize()
            
            with torch.no_grad():
                # 더 보수적인 생성 설정
                output_ids = self.llm.generate(
                    **inputs,
                    max_new_tokens=100,  # 더 짧게 설정
                    num_beams=3,  # beam 수 줄임
                    do_sample=False,
                    pad_token_id=self.tokenizer.eos_token_id,
                    early_stopping=True
                )
            
            # 입력 길이만큼 제거하여 새로 생성된 토큰만 디코딩
            generated_tokens = output_ids[0][inputs['input_ids'].shape[1]:]
            output = self.tokenizer.decode(generated_tokens, skip_special_tokens=True)
            
            # 메모리 정리
            del inputs, output_ids, generated_tokens
            gc.collect()
            if torch.cuda.is_available():
                torch.cuda.empty_cache()
            
            return output.strip() if output.strip() else "답변을 생성할 수 없습니다."
            
        except RuntimeError as e:
            if "CUDA" in str(e):
                print(f"CUDA 오류 발생: {e}")
                print("GPU 메모리를 정리하고 재시도합니다...")
                
                # 메모리 정리
                gc.collect()
                if torch.cuda.is_available():
                    torch.cuda.empty_cache()
                    torch.cuda.synchronize()
                
                return "⚠️ GPU 메모리 부족으로 답변 생성에 실패했습니다. 질문을 더 간단하게 해주세요."
            else:
                return f"⚠️ 답변 생성 실패: {e}"
        except Exception as e:
            return f"⚠️ 예상치 못한 오류: {e}"

    def chat(self):
        print("\n=== SLM & RAG 기반 대화형 챗봇 시작 ===")
        print("문서 내용 기반 검색 및 자연어 응답을 지원합니다. '종료' 입력 시 종료됩니다.")
        print("----------------------------------------------------")

        while True:
            try:
                user_input = input("\n어떤 주제에 대해 궁금하신가요?: ").strip()

                if user_input.lower() in ['종료', 'quit', 'exit']:
                    print("챗봇을 종료합니다.")
                    break

                keywords = self.extract_keywords(user_input)
                matched_docs = self.filter_docs_by_keywords(keywords)

                if not matched_docs:
                    print("해당 키워드와 관련된 문서를 찾을 수 없습니다. 다른 키워드를 입력해 주세요.")
                    continue

                print(f"\n[{', '.join(keywords)}] 키워드가 포함된 문서 일부를 검색했습니다.")
                print("관련 내용 중에서 질문을 해보세요.")

                while True:
                    followup = input("궁금한 내용을 입력해주세요: ").strip()
                    if followup.lower() in ['종료', 'quit', 'exit']:
                        print("챗봇을 종료합니다.")
                        return

                    print("\n문서 검색 중...")
                    results = self.db.similarity_search(followup, k=3)

                    if not results:
                        print("관련 내용을 찾을 수 없습니다. 다시 입력해주세요.")
                        continue

                    context = "\n\n".join(doc.page_content.strip() for doc in results)
                    answer = self.generate_answer(context, followup)

                    print("\n--- 답변 ---")
                    print(answer)
                    print("\n출처:")
                    for i, doc in enumerate(results, 1):
                        metadata = doc.metadata
                        print(f"- 문서{i}: {metadata.get('source', '알 수 없음')} (페이지: {metadata.get('page', '?')})")
                    print("\n다른 질문이 있으시면 계속 입력해주세요. 종료하려면 '종료'를 입력하세요.")
                    break
                    
            except KeyboardInterrupt:
                print("\n챗봇이 중단되었습니다.")
                break
            except Exception as e:
                print(f"오류 발생: {e}")
                print("계속 진행하려면 아무 키나 누르세요...")
                continue

    def cleanup(self):
        """메모리 정리 함수"""
        if hasattr(self, 'llm'):
            del self.llm
        if hasattr(self, 'tokenizer'):
            del self.tokenizer
        gc.collect()
        if torch.cuda.is_available():
            torch.cuda.empty_cache()


if __name__ == "__main__":
    chatbot = None
    try:
        chatbot = InteractiveRAGChatbot()
        chatbot.chat()
    except KeyboardInterrupt:
        print("\n챗봇이 중단되었습니다.")
    except Exception as e:
        print(f"심각한 오류 발생: {e}")
        print("FAISS 데이터베이스 또는 로컬 LLM 모델 구성을 확인해주세요.")
    finally:
        if chatbot:
            chatbot.cleanup()
        print("리소스 정리 완료.")
```

```python
import os
import re
import torch
import gc
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from peft import PeftModel, PeftConfig


class InteractiveRAGChatbot:
    def __init__(self, faiss_path="embed_faiss"):
        if torch.cuda.is_available():
            torch.cuda.empty_cache()
            torch.cuda.synchronize()
        
        print("임베딩 모델 로딩 중...")
        self.embedding_model = HuggingFaceEmbeddings(
            model_name="sentence-transformers/all-MiniLM-L6-v2",
            model_kwargs={'device': 'cpu'}
        )

        print("FAISS 벡터 데이터베이스 로딩 중...")
        self.db = FAISS.load_local(faiss_path, self.embedding_model, allow_dangerous_deserialization=True)

        print("Flan-T5 모델 + LoRA 설정 중...")
        base_model = "google/flan-t5-base"
        peft_model_path = "slm_lora"

        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        print(f"사용 디바이스: {self.device}")
        
        # LoRA 모델 존재 여부 확인
        self.check_lora_model(peft_model_path)
        
        # 토크나이저 로딩
        self.tokenizer = AutoTokenizer.from_pretrained(base_model)
        
        # EOS 토큰 확인 및 설정
        if self.tokenizer.eos_token is None:
            self.tokenizer.eos_token = self.tokenizer.pad_token
        if self.tokenizer.pad_token is None:
            self.tokenizer.pad_token = self.tokenizer.eos_token
        
        # 기본 모델 로딩
        try:
            print("기본 Flan-T5 모델 로딩 중...")
            if torch.cuda.is_available():
                self.base_model = AutoModelForSeq2SeqLM.from_pretrained(
                    base_model, 
                    torch_dtype=torch.float16,
                    low_cpu_mem_usage=True,
                    device_map="auto"
                )
            else:
                self.base_model = AutoModelForSeq2SeqLM.from_pretrained(
                    base_model,
                    torch_dtype=torch.float32
                )
                self.base_model = self.base_model.to(self.device)
            
            print("✅ 기본 모델 로딩 완료")
            
            # LoRA 모델 적용
            try:
                print("LoRA 어댑터 적용 중...")
                self.llm = PeftModel.from_pretrained(self.base_model, peft_model_path)
                
                # LoRA 모델이 제대로 로드되었는지 확인
                self.verify_lora_loading()
                
                print("✅ LoRA 모델 적용 완료")
                self.using_lora = True
                
            except Exception as e:
                print(f"❌ LoRA 모델 로딩 실패: {e}")
                print("기본 모델만 사용합니다.")
                self.llm = self.base_model
                self.using_lora = False
            
            self.llm.eval()
            
            gc.collect()
            if torch.cuda.is_available():
                torch.cuda.empty_cache()
                
        except Exception as e:
            print(f"모델 로딩 중 오류 발생: {e}")
            raise e
        
        print("모델 로딩 완료!")

    def check_lora_model(self, peft_model_path):
        """LoRA 모델 파일 존재 여부 확인"""
        print(f"\n=== LoRA 모델 파일 확인 ===")
        print(f"LoRA 경로: {peft_model_path}")
        
        if not os.path.exists(peft_model_path):
            print(f"❌ LoRA 디렉토리가 존재하지 않습니다: {peft_model_path}")
            return False
        
        # 필수 파일들 확인
        required_files = [
            "adapter_config.json",
            "adapter_model.bin"  # 또는 adapter_model.safetensors
        ]
        
        missing_files = []
        for file in required_files:
            file_path = os.path.join(peft_model_path, file)
            if os.path.exists(file_path):
                size = os.path.getsize(file_path)
                print(f"✅ {file}: {size:,} bytes")
            else:
                # .safetensors 버전도 확인
                safetensors_file = file.replace(".bin", ".safetensors")
                safetensors_path = os.path.join(peft_model_path, safetensors_file)
                if os.path.exists(safetensors_path):
                    size = os.path.getsize(safetensors_path)
                    print(f"✅ {safetensors_file}: {size:,} bytes")
                else:
                    missing_files.append(file)
                    print(f"❌ {file}: 파일이 없습니다")
        
        # 전체 디렉토리 내용 출력
        print(f"\n{peft_model_path} 디렉토리 내용:")
        try:
            for item in os.listdir(peft_model_path):
                item_path = os.path.join(peft_model_path, item)
                if os.path.isfile(item_path):
                    size = os.path.getsize(item_path)
                    print(f"  📄 {item}: {size:,} bytes")
                else:
                    print(f"  📁 {item}/")
        except Exception as e:
            print(f"디렉토리 읽기 실패: {e}")
        
        return len(missing_files) == 0

    def verify_lora_loading(self):
        """LoRA가 제대로 로드되었는지 확인"""
        print(f"\n=== LoRA 로딩 상태 확인 ===")
        
        # PeftModel 속성 확인
        if hasattr(self.llm, 'peft_config'):
            print("✅ peft_config 존재")
            for key, config in self.llm.peft_config.items():
                print(f"  📋 {key}: {config}")
        else:
            print("❌ peft_config 없음")
        
        # 어댑터 모듈 확인
        if hasattr(self.llm, 'base_model'):
            print("✅ base_model 래퍼 존재")
        else:
            print("❌ base_model 래퍼 없음")
        
        # LoRA 가중치가 있는 모듈 찾기
        lora_modules = []
        for name, module in self.llm.named_modules():
            if 'lora' in name.lower():
                lora_modules.append(name)
        
        if lora_modules:
            print(f"✅ LoRA 모듈 발견: {len(lora_modules)}개")
            for i, module_name in enumerate(lora_modules[:5]):  # 처음 5개만 출력
                print(f"  🔧 {module_name}")
            if len(lora_modules) > 5:
                print(f"  ... 및 {len(lora_modules) - 5}개 더")
        else:
            print("❌ LoRA 모듈을 찾을 수 없습니다")
        
        return len(lora_modules) > 0

    def test_lora_vs_base(self):
        """LoRA 모델과 기본 모델 비교 테스트"""
        print("\n=== LoRA vs 기본 모델 비교 테스트 ===")
        
        test_prompt = "Answer the question: What is machine learning? Answer:"
        
        inputs = self.tokenizer(
            test_prompt,
            return_tensors="pt",
            truncation=True,
            max_length=100
        )
        inputs = {k: v.to(self.device) for k, v in inputs.items()}
        
        # 1. 현재 모델 (LoRA 적용된) 테스트
        print("1. 현재 모델 (LoRA) 테스트:")
        try:
            with torch.no_grad():
                output_ids = self.llm.generate(
                    **inputs,
                    max_new_tokens=30,
                    do_sample=False,
                    pad_token_id=self.tokenizer.pad_token_id
                )
            
            full_output = self.tokenizer.decode(output_ids[0], skip_special_tokens=True)
            generated_part = self.tokenizer.decode(
                output_ids[0][inputs['input_ids'].shape[1]:], 
                skip_special_tokens=True
            )
            print(f"  전체 출력: '{full_output}'")
            print(f"  생성 부분: '{generated_part}'")
            
        except Exception as e:
            print(f"  ❌ 실패: {e}")
        
        # 2. 기본 모델과 비교 (LoRA 사용 시에만)
        if self.using_lora and hasattr(self, 'base_model'):
            print("\n2. 기본 모델 (LoRA 없음) 테스트:")
            try:
                with torch.no_grad():
                    base_output_ids = self.base_model.generate(
                        **inputs,
                        max_new_tokens=30,
                        do_sample=False,
                        pad_token_id=self.tokenizer.pad_token_id
                    )
                
                base_full_output = self.tokenizer.decode(base_output_ids[0], skip_special_tokens=True)
                base_generated_part = self.tokenizer.decode(
                    base_output_ids[0][inputs['input_ids'].shape[1]:], 
                    skip_special_tokens=True
                )
                print(f"  전체 출력: '{base_full_output}'")
                print(f"  생성 부분: '{base_generated_part}'")
                
                # 비교
                if generated_part != base_generated_part:
                    print("✅ LoRA 모델이 기본 모델과 다른 결과를 생성합니다 (LoRA 작동 중)")
                else:
                    print("❌ LoRA 모델과 기본 모델이 같은 결과를 생성합니다 (LoRA 미작동 의심)")
                    
            except Exception as e:
                print(f"  ❌ 기본 모델 테스트 실패: {e}")

    def extract_keywords(self, text):
        keywords = re.findall(r'[가-힣a-zA-Z0-9]+', text)
        return list(set(k for k in keywords if len(k) >= 2))

    def filter_docs_by_keywords(self, keywords):
        all_docs = self.db.similarity_search("", k=1000)
        matched = [doc for doc in all_docs if any(k in doc.page_content for k in keywords)]
        return matched

    def generate_answer(self, context, question, debug=False):
        context = context[:600]

        # LoRA 튜닝된 모델에 맞는 프롬프트 (학습 시 사용한 형식과 유사하게)
        prompts = [
            # 한국어 문서 기반 QA 형식 (일반적인 LoRA 튜닝 형식)
            f"다음 문서를 참고하여 질문에 답하세요.\n\n문서:\n{context}\n\n질문: {question}\n\n답변:",
            
            # 간단한 형식
            f"문서: {context}\n질문: {question}\n답변:",
            
            # 영어 명령 + 한국어 답변
            f"Answer in Korean based on the context.\nContext: {context}\nQuestion: {question}\nAnswer:",
            
            # 요약 형식
            f"다음 내용을 바탕으로 질문에 대해 3줄 이내로 답변하세요.\n\n내용: {context}\n\n질문: {question}\n\n답변:",
        ]

        for i, prompt in enumerate(prompts):
            if debug:
                print(f"\n--- 템플릿 {i+1} 시도 ---")
                print(f"LoRA 사용 여부: {self.using_lora}")
                print(f"프롬프트 길이: {len(prompt)} 문자")
            
            try:
                inputs = self.tokenizer(
                    prompt, 
                    return_tensors="pt", 
                    truncation=True, 
                    max_length=400,
                    padding=False
                )
                
                inputs = {k: v.to(self.device) for k, v in inputs.items()}
                input_length = inputs['input_ids'].shape[1]
                
                with torch.no_grad():
                    output_ids = self.llm.generate(
                        **inputs,
                        max_new_tokens=100,
                        min_length=input_length + 5,
                        do_sample=True,
                        temperature=0.7,
                        top_p=0.9,
                        pad_token_id=self.tokenizer.pad_token_id,
                        eos_token_id=self.tokenizer.eos_token_id,
                        early_stopping=False,
                        repetition_penalty=1.1
                    )
                
                if output_ids.shape[1] > input_length:
                    generated_tokens = output_ids[0][input_length:]
                    output = self.tokenizer.decode(generated_tokens, skip_special_tokens=True)
                    
                    if debug:
                        print(f"생성된 답변: '{output}'")
                    
                    del inputs, output_ids, generated_tokens
                    gc.collect()
                    if torch.cuda.is_available():
                        torch.cuda.empty_cache()
                    
                    clean_output = output.strip()
                    if clean_output and len(clean_output) > 1:
                        return clean_output
                
            except Exception as e:
                if debug:
                    print(f"템플릿 {i+1} 실패: {e}")
                continue
        
        status = "LoRA 모델" if self.using_lora else "기본 모델"
        return f"답변을 생성할 수 없습니다. ({status} 사용 중)"

    def chat(self):
        print("\n=== SLM & RAG 기반 대화형 챗봇 시작 ===")
        print(f"LoRA 모델 사용: {'✅ 예' if self.using_lora else '❌ 아니오 (기본 모델만 사용)'}")
        
        # LoRA vs 기본 모델 비교 테스트
        self.test_lora_vs_base()
        
        print("\n문서 내용 기반 검색 및 자연어 응답을 지원합니다.")
        print("디버그 모드: 'debug on', 종료: '종료'")
        print("----------------------------------------------------")
        
        debug_mode = False

        while True:
            try:
                user_input = input("\n어떤 주제에 대해 궁금하신가요?: ").strip()

                if user_input.lower() in ['종료', 'quit', 'exit']:
                    break
                    
                if user_input.lower() == 'debug on':
                    debug_mode = True
                    print("디버그 모드 활성화")
                    continue
                    
                if user_input.lower() == 'debug off':
                    debug_mode = False
                    print("디버그 모드 비활성화")
                    continue

                keywords = self.extract_keywords(user_input)
                matched_docs = self.filter_docs_by_keywords(keywords)

                if not matched_docs:
                    print("관련 문서를 찾을 수 없습니다.")
                    continue

                print(f"[{', '.join(keywords)}] 키워드 관련 문서를 찾았습니다.")

                while True:
                    followup = input("구체적인 질문을 입력하세요: ").strip()
                    if followup.lower() in ['종료', 'quit', 'exit']:
                        return

                    results = self.db.similarity_search(followup, k=3)
                    if not results:
                        print("관련 내용을 찾을 수 없습니다.")
                        continue

                    context = "\n\n".join(doc.page_content.strip() for doc in results)
                    answer = self.generate_answer(context, followup, debug=debug_mode)

                    print(f"\n--- 답변 ({'LoRA' if self.using_lora else '기본'} 모델) ---")
                    print(answer)
                    print("\n출처:")
                    for i, doc in enumerate(results, 1):
                        metadata = doc.metadata
                        print(f"- 문서{i}: {metadata.get('source', '?')} (페이지: {metadata.get('page', '?')})")
                    break
                    
            except KeyboardInterrupt:
                break
            except Exception as e:
                print(f"오류: {e}")
                continue

    def cleanup(self):
        if hasattr(self, 'llm'):
            del self.llm
        if hasattr(self, 'base_model'):
            del self.base_model
        if hasattr(self, 'tokenizer'):
            del self.tokenizer
        gc.collect()
        if torch.cuda.is_available():
            torch.cuda.empty_cache()


if __name__ == "__main__":
    chatbot = None
    try:
        chatbot = InteractiveRAGChatbot()
        chatbot.chat()
    except Exception as e:
        print(f"오류 발생: {e}")
    finally:
        if chatbot:
            chatbot.cleanup()
```

```python
import os
import re
import json
import pickle
from typing import List, Dict, Optional, Tuple
from datetime import datetime
import numpy as np

# LangChain imports
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.schema import Document
from langchain.memory import ConversationBufferWindowMemory
from langchain.chains import ConversationalRetrievalChain
from langchain.llms import HuggingFacePipeline
from langchain.prompts import PromptTemplate
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.memory import ConversationBufferMemory

# Transformers and ML libraries
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch
from sentence_transformers import SentenceTransformer

# Text processing
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import WordNetLemmatizer

# Download required NLTK data
try:
    nltk.data.find('tokenizers/punkt')
    nltk.data.find('corpora/stopwords')
    nltk.data.find('corpora/wordnet')
except LookupError:
    nltk.download('punkt')
    nltk.download('stopwords')
    nltk.download('wordnet')

class DocumentPreprocessor:
    def clean_junk_content(self, text: str) -> str:
        """정크 콘텐츠 제거"""
        # 한글, 영어, 숫자, 기본 구두점만 허용하고 나머지 제거 (특수기호는 일부 유지)
        text = re.sub(r'[^가-힣a-zA-Z0-9\s.,!?()\-]', ' ', text)
        
        # 연속된 공백 제거
        text = re.sub(r'\s+', ' ', text)
        
        # 1글자 단어 제거 (영어, 한글 모두)
        text = re.sub(r'\b\w{1}\b', '', text)
        
        # 숫자만 있는 줄 제거
        text = re.sub(r'^\d+$', '', text, flags=re.MULTILINE)
        
        # 페이지 번호, 헤더/푸터 패턴 제거 강화 (한글/영어 혼용 고려)
        text = re.sub(r'페이지\s*\d+|Page\s*\d+|^\s*-\s*\d+\s*-\s*$', '', text, flags=re.MULTILINE)
        
        return text.strip()

    
    def preprocess_text(self, text: str) -> str:
        """텍스트 전처리"""
        text = self.clean_junk_content(text)
        
        # 영어 대문자 모두 소문자로 변환
        text = text.lower()
        
        sentences = sent_tokenize(text)
        cleaned_sentences = []
        
        for sentence in sentences:
            if 10 <= len(sentence) <= 500:
                cleaned_sentences.append(sentence)
        
        return ' '.join(cleaned_sentences)
    
    def extract_keywords(self, text: str, max_keywords: int = 10) -> List[str]:
        """향상된 키워드 추출"""
        tokens = word_tokenize(text.lower())
        
        keywords = []
        for token in tokens:
            if (len(token) > 2 and
                token not in self.stop_words and
                token not in set(k.lower() for k in self.korean_stop_words) and
                not token.isdigit()):
                lemmatized = self.lemmatizer.lemmatize(token)
                keywords.append(lemmatized)
        
        from collections import Counter
        keyword_freq = Counter(keywords)
        return [kw for kw, freq in keyword_freq.most_common(max_keywords)]


def _hybrid_retrieval(self, query: str, db, k: int = 3) -> List[Document]:
        """하이브리드 검색 (의미 + 키워드)"""
        semantic_docs = db.similarity_search(query, k=k*2)
        
        # query 키워드 추출을 전처리 객체를 써서 개선
        preprocessor = DocumentPreprocessor()
        query_keywords = set(preprocessor.extract_keywords(query))
        
        scored_docs = []
        for doc in semantic_docs:
            doc_keywords = set(preprocessor.extract_keywords(doc.page_content))
            keyword_overlap = len(query_keywords.intersection(doc_keywords))
            scored_docs.append((doc, keyword_overlap))
        
        scored_docs.sort(key=lambda x: x[1], reverse=True)
        return [doc for doc, score in scored_docs[:k]]
    
class ConversationMemory:
    """대화 메모리 관리 클래스"""
    
    def __init__(self, max_history: int = 10):
        self.max_history = max_history
        self.conversation_history = []
        self.context_keywords = set()
        
    def add_exchange(self, question: str, answer: str, keywords: List[str] = None):
        """대화 교환 추가"""
        exchange = {
            'timestamp': datetime.now(),
            'question': question,
            'answer': answer,
            'keywords': keywords or []
        }
        
        self.conversation_history.append(exchange)
        
        # 키워드 업데이트
        if keywords:
            self.context_keywords.update(keywords)
        
        # 히스토리 길이 제한
        if len(self.conversation_history) > self.max_history:
            old_exchange = self.conversation_history.pop(0)
            # 오래된 키워드 제거 (간단한 방식)
            if old_exchange['keywords']:
                for kw in old_exchange['keywords']:
                    if kw in self.context_keywords:
                        self.context_keywords.discard(kw)
    
    def get_context(self) -> str:
        """현재 대화 컨텍스트 반환"""
        if not self.conversation_history:
            return ""
        
        context_parts = []
        for exchange in self.conversation_history[-3:]:  # 최근 3개 대화만
            context_parts.append(f"Q: {exchange['question']}")
            context_parts.append(f"A: {exchange['answer'][:100]}...")  # 답변은 100자만
        
        return "\n".join(context_parts)
    
    def get_relevant_keywords(self) -> List[str]:
        """관련 키워드 반환"""
        return list(self.context_keywords)

class NaturalLanguageGenerator:
    """자연스러운 문장 생성 클래스"""
    
    def __init__(self):
        self.templates = {
            'answer': [
                "검색된 문서에 따르면, {content}",
                "문서 내용을 바탕으로 설명드리면, {content}",
                "관련 자료에서 찾은 정보는 다음과 같습니다: {content}",
                "{content}"
            ],
            'no_result': [
                "죄송하지만 관련된 정보를 찾을 수 없습니다.",
                "해당 주제에 대한 문서를 찾지 못했습니다.",
                "요청하신 내용과 관련된 자료가 데이터베이스에 없는 것 같습니다."
            ],
            'clarification': [
                "좀 더 구체적인 질문을 해주시면 더 정확한 답변을 드릴 수 있습니다.",
                "질문을 다시 한번 명확히 해주실 수 있나요?",
                "어떤 부분에 대해 더 자세히 알고 싶으신가요?"
            ]
        }
    
    def generate_response(self, content: str, response_type: str = 'answer') -> str:
        """자연스러운 응답 생성"""
        if response_type in self.templates:
            import random
            template = random.choice(self.templates[response_type])
            if '{content}' in template:
                return template.format(content=content)
            else:
                return template
        return content

class AdvancedRAGChatbot:
    def __init__(self, faiss_path, embedding_model, use_memory=True):
        self.db = FAISS.load_local(faiss_path, HuggingFaceEmbeddings(model_name=embedding_model), allow_dangerous_deserialization=True)
        self.tokenizer = AutoTokenizer.from_pretrained("openlm-research/open_llama_3b_v2")
        self.model = AutoModelForCausalLM.from_pretrained(
            "lorapeft_flan_t5_lora",
            torch_dtype=torch.float16,
            device_map="auto"
        )
        self.text_splitter = RecursiveCharacterTextSplitter(chunk_size=350, chunk_overlap=50)
        self.use_memory = use_memory
        if use_memory:
            self.memory = ConversationBufferMemory(memory_key="chat_history", return_messages=True)

    def retrieve_relevant_chunks(self, query, k=3):
        docs = self.db.similarity_search(query, k=k)
        return docs

    def summarize_chunks(self, docs):
        summary = "\n".join([doc.page_content.strip().replace("\n", " ") for doc in docs])
        return summary[:1000]  # 길이 제한

    def generate_answer(self, query, summary):
        prompt = f"""사용자 질문: {query}
참고 문서 내용:
{summary}

위 내용을 바탕으로 3줄 이내로 자연스럽게 답변해줘.
"""
        inputs = self.tokenizer(prompt, return_tensors="pt", truncation=True, max_length=1024).to(self.model.device)
        with torch.no_grad():
            outputs = self.model.generate(**inputs, max_new_tokens=128, pad_token_id=self.tokenizer.eos_token_id)
        answer = self.tokenizer.decode(outputs[0], skip_special_tokens=True)
        return answer.split("\n")[-1].strip()

    def format_sources(self, docs):
        sources = set()
        for d in docs:
            meta = d.metadata
            source = f"{meta.get('source', '')} p.{meta.get('page', '?')} 문단 {meta.get('paragraph', '?')}"
            sources.add(source)
        return ", ".join(sources)

    def chat_session(self):
        print("💬 챗봇이 시작되었습니다. '종료'를 입력하면 끝납니다.")
        while True:
            user_input = input("사용자: ")
            if user_input.lower() in ["종료", "exit"]:
                print("챗봇을 종료합니다.")
                break
            try:
                docs = self.retrieve_relevant_chunks(user_input)
                summary = self.summarize_chunks(docs)
                answer = self.generate_answer(user_input, summary)
                sources = self.format_sources(docs)
                print(f"🤖 답변: {answer}\n📎 출처: {sources}")
            except Exception as e:
                print(f"⚠️ 오류 발생: {e}")

# 실행 함수
def run_advanced_chatbot():
    """고도화된 챗봇 실행"""
    try:
        # 기본 임베딩 모델 지정
        embedding_model = "sentence-transformers/all-MiniLM-L6-v2"
        
        chatbot = AdvancedRAGChatbot(
            faiss_path="embed_faiss",
            embedding_model=embedding_model,
            use_memory=True
        )
        
        chatbot.chat_session()
        
    except Exception as e:
        print(f"❌ 챗봇 초기화 실패: {e}")
        print("FAISS 데이터베이스가 올바르게 생성되었는지 확인해주세요.")

if __name__ == "__main__":
    run_advanced_chatbot()
```
